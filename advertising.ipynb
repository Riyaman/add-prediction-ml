```python
import pandas as pd
import numpy as np
from sklearn import metrics
import sklearn as sk
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import train_test_split
```

```python
adv = pd.read_csv('advertising.csv')
```

```python
adv
```

```python
adv.info()
```

```python
adv.describe()
```

```python
train,test = train_test_split(adv,test_size=0.25)
```

```python
train
```

```python
test
```

```python
train_X = train[['Daily Time Spent on Site','Age','Area Income','Daily Internet Usage']]
train_y = train[['Clicked on Ad']]
train_X
```

```python
train_y
```

```python
test_X = test[['Daily Time Spent on Site','Age','Area Income','Daily Internet Usage']]
test_y = test[['Clicked on Ad']]
test_X
```

```python
test_y
```

```python
model = svm.SVC()
```

```python
model.fit(train_X,train_y)
pred = model.predict(test_X)
```

```python
pred
```

```python
metrics.accuracy_score(pred,test_y)
```

```python
model = DecisionTreeClassifier()
model.fit(train_X,train_y)
pred = model.predict(test_X)
```

```python
metrics.accuracy_score(pred,test_y)
```

```python
model = KNeighborsClassifier()
model.fit(train_X,train_y)
pred = model.predict(test_X)
```

```python
metrics.accuracy_score(pred,test_y)
```

```python
model = RandomForestClassifier()
model.fit(train_X,train_y)
pred = model.predict(test_X)
```

```python
metrics.accuracy_score(pred,test_y)
```
